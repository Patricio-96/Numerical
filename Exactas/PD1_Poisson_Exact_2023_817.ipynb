```python
def poisson_2d_exacta(x,y):  
  N = 50
  u = 0
  for k in range(1,N+1):
    for l in range(1,N+1):
      # sea ukl el factor que multiplica al producto de senos
      ukl = 4*np.power(-1,k+l) / ( k*l*(k**2+l**2)*np.power(np.pi,4) ) 

      # acumulamos la suma en u para los valores de entrad (x,y)
      u = u + ukl*np.sin(k*np.pi*x)*np.sin(l*np.pi*y)

  return u
```

```python
poisson_2d_exacta(0.4,0.5)
```
